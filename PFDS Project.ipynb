# Data Visualisation using Various Graphs

# ENERGY CONSUMPTION

# Which city has the highest energy consumption??

```python
avg_energy_by_city = df.groupby('City')['Energy Consumption'].mean().sort_values()
avg_energy_by_city.plot(kind='bar', title=' Energy Consumption by City')
plt.xlabel('City')
plt.ylabel('Average Energy Consumption')
plt.show()
```

```python
data1=df
sns.barplot(x='City', y='Energy Consumption', data = data1)
plt.show()
```

# Which vehicle consumes high energy??

```python
avg_energy_by_city = df.groupby('Vehicle Type')['Energy Consumption'].mean().sort_values()
avg_energy_by_city.plot(kind='bar', title='Energy Consumption by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Energy Consumption')
plt.show()
```

# At what time the energy was consumed the most ??

```python
energy_by_hour = df.groupby('Hour Of Day')['Energy Consumption'].mean()
energy_by_hour.plot(kind='line', title='Energy Consumption Variation by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Energy Consumption')
plt.show()
```

# On which day energy was consumed the most??

```python
energy_by_hour = df.groupby('Day Of Week')['Energy Consumption'].mean()
energy_by_hour.plot(kind='line', title='Energy Consumption Variation by  Day Of Week')
plt.xlabel('Day Of Week')
plt.ylabel('Average Energy Consumption')
plt.show()
```

# In which weather the energy was consumed the most??

```python

sns.lineplot(x='Weather', y='Energy Consumption', data=df)
```
